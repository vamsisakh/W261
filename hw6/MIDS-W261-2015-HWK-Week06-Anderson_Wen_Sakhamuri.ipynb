### Answer:
First order necessary conditions for optimality states that the first derivate of a function,f(x) has a zero value at either the local maximum or local minimum of the function f(x). 

Mathematically, solving for $f^{'}(x=x^{*}) = 0$ gives us either the maxima or minima of a function. 

Second order necessary conditions for optimality states that the second derivate of a function,f(x) can tell us whether the inputs correspond to a local maximum or a local minimum of the function f(x).

$f^{''}(x=x^{*}) < 0 (\ x^{*} \ corresponds\ to\ a\ minima)$  
$f^{''}(x=x^{*}) > 0 (\ x^{*} \ corresponds\ to\ a\ maxima)$

Note that for a concave function , local maximum is the same as the global maximum and for a convex function, the local minimum is the same as the global minimum.

```python
import numpy as np

import matplotlib.pyplot as plt

#Plotting f(x) over [-6,6)
r = np.arange(-6,6,0.01)
out = []

for x in r:    
    y = x**3 - 12*(x**2) - 6
    if((x)==0):
        print y
    out.append(y)

#Index at max value
idx_max = out.index(max(out))

#Plot
plt.plot(r,out,label="f(x)")
plt.ylabel('Function (x^3 - 12*(x^2)-6)')
plt.axvline(x=r[idx_max],color = 'r')
plt.annotate('Maxima!!', xy=(r[idx_max],out[idx_max]), xytext=(1, -100),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.legend(loc='lower right')
plt.show()

#Plotting 1st derivative of f(x) over [-6,6)
out = []
for x in r:
    y = 3*(x**2) - 24*(x)
    out.append(y)

#Plot
plt.plot(r,out,label="First Derivate of f(x)")
plt.ylabel('First Derivative')
plt.axvline(x=r[idx_max],color='r')
plt.axhline(y=out[idx_max],color='g')
plt.annotate('1st derivate is zero (maybe maxima or minima)', xy=(r[idx_max],out[idx_max]), xytext=(1, 25),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.legend(loc='upper right')
plt.show()

#Plotting 2nd derivative of f(x) over [-6,6)
out = []
for x in r:
    y = 6*x - 24 
    out.append(y)

#Plot
plt.plot(r,out,label = 'Second Derivate of f(x)')
plt.ylabel('Second Derivative')
plt.axvline(x=r[idx_max],color='r')
plt.axhline(y=out[idx_max],color='g')
plt.annotate('2nd derivate is negative (maxima)', xy=(r[idx_max],out[idx_max]), xytext=(1, 1),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.legend(loc='lower right')
plt.show()

```

From first order optimality condition, we can see from the second plot above the point where a maxima or minima will occur.

From second order optimality condition, we can see from the third plot that the point (obtained from FOC) corresponds to a maxima since the second derivative at that point is negative.

*Taking x=1 as the first approximation(xt1) of a root of X^3 + 2x -4 = 0, use the Newton-Raphson method to calculate the second approximation (denoted as xt2) of this root. (Hint the solution is xt2=1.2)*


## Answer:

Newton Raphson method involves computing the tangent line to a curve at a initial point (in this case x = 1) and then recomputing the tangent line now at the x-intercept (obtained from the previous tangent line).

Equation to a line:

$f(x)=mx+c$    ->equation(1)

where m is the slope and c is the y-intercept.

Let our initial point be $x^{*}$ where we would compute the tangent line.

At $x^{*}$, the slope of the line is simply the derivative at $x_{*}$, which is $f^{'}(x^{*})$

The exact point where the tangent intersects the line is (x,y) = $(x^{*},f(x^{*}))$

And so the equation to the tangent line would be:

$f(x) = f^{'}(x_{*})x+c$  -> equation(2)

We know that (x,y) = $(x^{*},f(x^{*}))$ are points on equation(2)

Subsituting,

$f(x^{*}) = f^{'}(x^{*})x^{*} + c$

And so the y-intercept , $c = f(x^{*}) - f^{'}(x^{*})x^{*}$

Subsituting this value of the intercept into equation(2)...

$f(x) = f^{'}(x^{*})x+f(x^{*}) - f^{'}(x^{*})x^{*}$  -- equation(3)

Now , equatio(3) is the equation to our tangent line. We need to compute the x-intercept of this line
Setting $f(x)$ to 0 to compute the x-intercept

$f^{'}(x^{*})x = f^{'}(x^{*})x^{*} - f(x^{*}) $

And so , here is our update rule,

$x = (x^{*}) - \frac{f(x^{*})}{f^{'}(x^{*})}$

In this case, our first approximation is $x^{*}=1$

And so , $x =  1 - \frac{f(1)}{f^{'}(1)}$

$f(1) = -1$

$f^{'}(1) = 5 $

$x =  1 + \frac{1}{5}$

$x = 1 + 0.2 = 1.2 $

## Answer

An optimization problem is convex if its objective is a convex function, the inequality constraints $f_{j}$ are convex, and the equality constraints $h_{j}$ are affine.

$minimize_{x} f_{0}(x)$ (Convex Function)

such that 

$f_{i}(x) \leq 0$ (Convex sets)

$h_{i}(x) = 0$ (Affine)


**First Order condition for convexity:**

A real valued differentiable function is convex iff

$f(y) \geq f(x) + \nabla f(y)^{T}.(y-x)$

for all x,y $\epsilon R^{n}$

**Second Order condition for convexity:**

Convexity can make optimization in some sense "easier" than the general case; e.g., any local minimum must be a global minimum.

So in essence, we do not need to look at second order condition to determine maximum or minimum

*The learning objective function for weighted ordinary least squares (WOLS) (aka weight linear regression) is defined as follows:*

`0.5* sumOverTrainingExample i (weight_i * (W * X_i - y_i)^2)`

*Where training set consists of input variables X ( in vector form) and a target variable y, and W is the vector of coefficients for the linear regression model.*

*Derive the gradient for this weighted OLS by hand; showing each step and also explaining each step.*


### Answer:
Learning Objective Cost function 

$$J(W) = 0.5* \Sigma_{i} (weight_i * (W * X_i - y_i)^2)$$

$$\frac{\partial J(W)}{\partial W} = \frac{\partial (0.5* \Sigma_{i} (weight_i * (W * X_i - y_i)^2))}{\partial W}$$

$$\frac{\partial J(W)}{\partial W} = 2*0.5 * \Sigma_{i} [(weight_i * (W * X_i - y_i) * x^i_j]$$

In the above equation $x^i_j$ is the $j^{th}$ feature from the $i^{th}$ training example

And so the gradient for our objective function is:

$$\frac{\partial J(W)}{\partial W} = \Sigma_{i} [(weight_i * (W * X_i - y_i) * x^i_j]$$

**Update rules for gradient descent:**

$$w_j := w_j - \alpha * \frac{\partial J(W)}{\partial W}$$

where $w_j$ is the jth coefficient from the coeffcient vector W and $\alpha$ is the learning rate
